```python
def N_t(N,t, nuclearData):
    iso_names = list(nuclearData.keys())
    lam_N = np.zeros(len(nuclearData))
    for i in nuclearData:
        lam = np.log(2)/nuclearData[i][0] #half-life equation
        iso_index = iso.index(i) #half-life calculation for all identified daughters
        lam_N[iso_index] += - lam * N[iso_index] #vector to store all of the decay rates
        for p in nuclearData[i][1]: #a loop to reference the daughters
            iso_prod = p[0] #pulls the daughter's isotope name
            Branch_Ratio = p[2] #this references the branching ratio's probability of occurring
            prod_index = iso.index(iso_prod) #relates daughter's name and half-life
            lam_N[prod_index] += Branch_Ratio * lam * N[iso_index] #gives the gain rate of the daughters
    return lam_N
```

```python
from scipy.integrate import odeint
N0 = np.zeros(len(nuclearData))
N0[0]=1
t = np.linspace(1,86400, 1000)
N = odeint(N_t,N0,t, args=(nuclearData,))
iso = list(nuclearData.keys()) #creates a list to loop through

plt.figure(dpi=300)
for i in range(N.shape[1]):
    plt.plot(t, N[:,i], label=iso[i])

plt.legend()    
plt.xlabel("Time's'")
plt.ylabel('Relative Abundance')
plt.title('Problem 5')
plt.show()
```
